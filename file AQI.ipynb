```python
def cal_CO2i(co2):
    co2i = 0
    if co2 <= 1000:
        co2i = co2 * 50 / 1000
    elif co2 > 1000 and co2 <= 2000:
        co2i = 50 + (co2 - 1000) * 50 / 1000
    elif co2 > 2000 and co2 <= 5000:
        co2i = 100 + (co2 - 2000) * 100 / 3000
    elif co2 > 5000:
        co2i = 200 + (co2 - 5000) * 100 / 5000
    return co2i

df['CO2i'] = df['co2'].apply(cal_CO2i)
```

```python
def cal_NO2i(no2):
    no2i = 0
    if no2 <= 40:
        no2i = no2 * 50 / 40
    elif no2 > 40 and no2 <= 80:
        no2i = 50 + (no2 - 40) * 50 / 40
    elif no2 > 80 and no2 <= 180:
        no2i = 100 + (no2 - 80) * 100 / 100
    elif no2 > 180 and no2 <= 280:
        no2i = 200 + (no2 - 180) * 100 / 100
    elif no2 > 280 and no2 <= 400:
        no2i = 300 + (no2 - 280) * 100 / 120
    elif no2 > 400:
        no2i = 400 + (no2 - 400) * 100 / 400
    return no2i

df["NO2i"] = df["no2"].apply(cal_NO2i)
```

```python
def cal_COi(co):
    coi = 0
    if co <= 4.4:
        coi = co * 50 / 4.4
    elif co > 4.4 and co <= 9.4:
        coi = 50 + (co - 4.4) * 50 / 5.0
    elif co > 9.4 and co <= 12.4:
        coi = 100 + (co - 9.4) * 100 / 3.0
    elif co > 12.4 and co <= 15.4:
        coi = 200 + (co - 12.4) * 100 / 3.0
    elif co > 15.4 and co <= 30.4:
        coi = 300 + (co - 15.4) * 100 / 15.0
    elif co > 30.4:
        coi = 400 + (co - 30.4) * 100 / 30.0
    return coi

df["COi"] = df["co"].apply(cal_COi)
```

```python
def cal_Benzenei(benzene):
    benzenei = 0
    if benzene <= 3:
        benzenei = benzene * 50 / 3
    elif benzene > 3 and benzene <= 10:
        benzenei = 50 + (benzene - 3) * 50 / 7
    elif benzene > 10 and benzene <= 20:
        benzenei = 100 + (benzene - 10) * 100 / 10
    elif benzene > 20 and benzene <= 50:
        benzenei = 200 + (benzene - 20) * 100 / 30
    elif benzene > 50 and benzene <= 100:
        benzenei = 300 + (benzene - 50) * 100 / 50
    elif benzene > 100:
        benzenei = 400 + (benzene - 100) * 100 / 100
    return benzenei

df["Benzenei"] = df["benzene"].apply(cal_Benzenei)
```

```python

def cal_Toluenei(toluene):
    toluenei = 0
    if toluene <= 10:
        toluenei = toluene * 50 / 10
    elif toluene > 10 and toluene <= 50:
        toluenei = 50 + (toluene - 10) * 50 / 40
    elif toluene > 50 and toluene <= 100:
        toluenei = 100 + (toluene - 50) * 100 / 50
    elif toluene > 100 and toluene <= 200:
        toluenei = 200 + (toluene - 100) * 100 / 100
    elif toluene > 200 and toluene <= 500:
        toluenei = 300 + (toluene - 200) * 100 / 300
    elif toluene > 500:
        toluenei = 400 + (toluene - 500) * 100 / 100
    return toluenei

df["Toluenei"] = df["toluene"].apply(cal_Toluenei)
```

```python

def cal_PM25i(pm25):
    pm25i = 0
    if pm25 <= 30:
        pm25i = pm25 * 50 / 30
    elif pm25 > 30 and pm25 <= 60:
        pm25i = 50 + (pm25 - 30) * 50 / 30
    elif pm25 > 60 and pm25 <= 90:
        pm25i = 100 + (pm25 - 60) * 100 / 30
    elif pm25 > 90 and pm25 <= 120:
        pm25i = 200 + (pm25 - 90) * 100 / 30
    elif pm25 > 120 and pm25 <= 250:
        pm25i = 300 + (pm25 - 120) * 100 / 130
    elif pm25 > 250:
        pm25i = 400 + (pm25 - 250) * 100 / 130
    return pm25i

df["PM25i"] = df["pm25"].apply(cal_PM25i)
```

```python
import numpy as np
import pandas as pd


def get_AQI_bucket(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Moderate"
    elif x <= 200:
        return "Unhealthy for Sensitive Groups"
    elif x <= 300:
        return "Unhealthy"
    elif x <= 400:
        return "Very Unhealthy"
    elif x > 400:
        return "Hazardous"
    else:
        return np.nan

# Calculate Checks, excluding PM10_SubIndex as it's missing
df["Checks"] =  (df["CO2i"] > 0).astype(int) + \
                (df["NO2i"] > 0).astype(int) + \
                (df["COi"] > 0).astype(int) + \
                (df["Benzenei"] > 0).astype(int) + \
                (df["Toluenei"] > 0).astype(int) + \
                (df["PM25i"] > 0).astype(int)

df["AQI"] = round(df[['CO2i', 'NO2i', 'COi', 'Benzenei', 'Toluenei', 'PM25i']].max(axis = 1))
df.loc[df["PM25i"] <= 0, "AQI"] = np.nan
df.loc[df.Checks < 3, "AQI"] = np.nan

df["AQI_Range"] = df["AQI"].apply(lambda x: get_AQI_bucket(x))
df[~df.AQI.isna()].head(13)
```
